```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([23, 34, 42, 49, 55, 61, 68, 73, 78, 82])
y = np.array([25, 29, 33, 37, 44, 47, 57, 62, 67, 71])
z = np.array([87, 82, 78, 73, 67, 63, 55, 51, 47, 39])
```

# Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5)) 
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'Red', color = 'Black', ls = '--', ms = 5, label = 2024);
plt.plot(x, z, marker = 'o', mfc = 'Yellow', color = 'Black', ls = '--', ms = 5, label = 2023);
plt.grid()
plt.legend();

# markers => >, <, ^, H, O,*
```

```python
plt.plot(x, z);
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5)) 
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'Red', color = 'Black', ms = 5);
plt.plot(x, z, marker = 'o', mfc = 'Yellow', color = 'Black', ms = 5);
plt.grid()
```

```python
# subplot
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5)) 
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'Red', color = 'Black', ls = '--', ms = 5);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5)) 
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'Yellow', color = 'Black', ls = '--', ms = 5);
plt.grid()
```

# Bar Plot

```python

courses = ['java', 'python', 'kotlin', 'dart', 'R']
student = [67, 63, 59, 55, 47]
```

# Pie Chart

```python
colors = ['pink', 'orange', 'skyblue', 'green', 'purple']
plt.pie(student, labels = courses, autopct = '%.2f%%', labeldistance = 1.1, startangle = 90, shadow = True, explode = [0.3, 0, 0, 0, 0],colors = colors);
```

```python
plt.figure(figsize =(18,14))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker = 'o', mfc = 'red', color = 'green' ,ls ='--', ms = 8 ,label = '2023');
plt.plot(x,z,marker = 'o', mfc = 'blue', color = 'green' ,ls ='--', ms = 8,label = '2024');
plt.grid()     

plt.subplot(2,2,2)
plt.title('Bar Plot')
plt.xlabel('courses')
plt.ylabel('student')
plt.bar(courses ,student ,color = 'Black', edgecolor = 'pink' , width = 0.5 , linewidth = 3,alpha = 0.5);
plt.grid()

plt.subplot(2,2,3)
plt.title('scatter plot')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.scatter(df['Age'],df['EstimatedSalary'], c = df['Purchased'] );

plt.subplot(2,2,4)
plt.title('pie chart')
colors = ['pink' , 'orange' , 'blue' , 'green' , 'purple']
plt.pie(student, labels = courses , autopct = '%.2f%%' , labeldistance = 1.2 , startangle = 90 , shadow = True , explode =[0.2,0,0,0,0],
        colors = colors);

plt.savefig('myplot.png')
```

# box plot

```python
marks = [67, 65, 55, 59, 69, 72, 51, 70, 53, 98, 23]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3,q1
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
ub = q3 +(1.5 * iqr)
lb = q1 -(1.5 * iqr)
```
